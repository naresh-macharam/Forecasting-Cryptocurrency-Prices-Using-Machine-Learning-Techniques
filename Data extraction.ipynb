##Alternative.me fear & greed data from past 5 years.

At the time of running code data retrieved from sep-11-2018 to sep-9-2023

```python
import requests
import pandas as pd
```

```python
# Define the API endpoint URL for historical data
url = "https://api.alternative.me/fng/"

# Define the number of days for historical data (1825 for 5 years)
days = 1825

# Send a GET request to the API with the days parameter
response = requests.get(f"{url}?limit={days}")

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()

    # Extract the historical Fear and Greed Index data
    historical_data = data['data']

    # Create a DataFrame from the historical data
    df = pd.DataFrame(historical_data)

    # Save the data to a CSV file
    df.to_csv('fear_and_greed_indexx.csv', index=False)

    print("Historical Fear and Greed Index data saved to 'fear_and_greed_indexx.csv'.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

##Crypto data from yahoo finance OHLC hourly data for 2 years
At the time of running code data retrieved from aug-7-2021 to oct-01-2022

```python
import yfinance as yf
import datetime as dt
import pandas as pd


def get_crypto_hourly_data(ticker_symbols):
    # Calculate the date range for the last 729 days (2 years)
    end_date = dt.date.today()
    start_date = end_date - dt.timedelta(days=729)

    crypto_data = {}
    for symbol in ticker_symbols:
        # Download the hourly data for each cryptocurrency using yfinance
        data = yf.download(symbol, start=start_date, end=end_date, interval='1h')
        crypto_data[symbol] = data

    return crypto_data



# List of ticker symbols for the selected cryptocurrencies
tickers = ['BTC-USD', 'ETH-USD', 'USDT-USD', 'USDC-USD', 'BNB-USD']

# Call the function to get hourly data for each cryptocurrency
crypto_hourly_data = get_crypto_hourly_data(tickers)

# Save each cryptocurrency's data to individual CSV files
for symbol, data in crypto_hourly_data.items():
      csv_file_path = f"{symbol.lower()}_hourly_data.csv"
      data.to_csv(csv_file_path)
      print(f"{symbol} hourly data saved to {csv_file_path}")
```
