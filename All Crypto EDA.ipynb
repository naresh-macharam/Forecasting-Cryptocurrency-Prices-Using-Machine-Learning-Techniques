```python
df_tether = pd.read_csv("./usdt-usd_hourly_data.csv")
df_usdc = pd.read_csv("./usdc-usd_hourly_data.csv")
df_btc = pd.read_csv("./btc-usd_hourly_data.csv")
df_eth = pd.read_csv("./eth-usd_hourly_data.csv")
df_bnb = pd.read_csv("./bnb-usd_hourly_data.csv")
```

```python




# Convert 'Date' column to datetime type
for df in [df_btc, df_eth, df_bnb, df_tether, df_usdc]:
    df['Date'] = pd.to_datetime(df['Datetime'])

# Create a single plot with multiple lines
plt.figure(figsize=(12, 8))

plt.plot(df_btc['Date'], df_btc['Close'], label='BTC', marker='o', linestyle='-')
plt.plot(df_eth['Date'], df_eth['Close'], label='ETH', marker='o', linestyle='-')
plt.plot(df_bnb['Date'], df_bnb['Close'], label='BNB', marker='o', linestyle='-')
plt.plot(df_tether['Date'], df_tether['Close'], label='Tether', marker='o', linestyle='-')
plt.plot(df_usdc['Date'], df_usdc['Close'], label='USD coin', marker='o', linestyle='-')

plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.title('Closing Prices of cryptos')
plt.legend()
plt.grid(True)
plt.show()
```
